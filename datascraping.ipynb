```python
import pandas as pd
```

```python
tickers = pd.read_csv('nasdaq_screener_1718368595920.csv')
```

```python
tickers = tickers.sort_values(by = ['Market Cap'], ascending = False)
```

```python
tickers = tickers['Symbol'][:200]
```

```python
tickers
```

```python
import yfinance as yahooFinance
import datetime
test_date = datetime.datetime.strptime("2012-01-01", "%Y-%m-%d")
startDate = datetime.datetime(2012, 1, 1)
endDate = datetime.datetime(2023, 12, 31)


df = pd.DataFrame()

for tk in tickers:
    GetFacebookInformation = yahooFinance.Ticker(tk)
    # print(GetFacebookInformation)
 
    # Let us  get historical stock prices for Facebook 
    # covering the past few years.
    # max->maximum number of daily prices available 
    # for Facebook.
    # Valid options are 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 
    # 5y, 10y and ytd.
    tkf = (GetFacebookInformation.history(start = startDate, end = endDate)['Open'])
    if(tkf.shape[0] != 3018):
        continue
    tkf = pd.DataFrame(tkf)
    tkf.index = (tkf.index.date)
    tkf.columns = {tk}
    if(tk == 'AAPL'):
        df = tkf
    else:
        df = pd.merge(df, tkf, left_index = True, right_index = True)
```
